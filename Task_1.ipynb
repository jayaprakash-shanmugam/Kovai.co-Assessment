```python
data1['date_added']=pd.to_datetime(data1['date_added'])
data1['date_added']
```

```python
data1.dtypes
```

```python
#converting date object to date
data1['date']=pd.DatetimeIndex(data1['date_added']).day
data1['month']=pd.DatetimeIndex(data1['date_added']).month
data1['year']=pd.DatetimeIndex(data1['date_added']).year
```

```python
#formating date_added column
```

```python
data1.head()
```

```python
data1.sort_values(by='date_added',ascending=False)
```

```python
# total movies by director
```

```python
by_dir=data1.groupby('director')['title'].count().sort_values(ascending=False).reset_index()
by_dir.head()
```

**Director Raúl Campos, Jan Suter has the most movied directed**

```python
# Total movies released by year
```

```python
by_dir=data1.groupby('release_year')['title'].count().sort_values(ascending=False).reset_index()
by_dir.head(5)
```

```python
plt.figure(figsize=(5,3))
sns.lineplot(data=by_dir.nlargest(10,'title'), x='release_year', y='title')
plt.title('Total reviews for each genre')
plt.xlabel ('release_year', fontsize=16)
plt.ylabel ('Total title', fontsize=16)
plt.show()
```

**2017 has the most movie releases**
